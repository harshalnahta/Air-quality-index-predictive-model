```python
import pandas as pd
data = pd.read_csv('aqidata.csv')
data.columns
data.head()
```

```python
from sklearn.linear_model import LinearRegression
predict_model = LinearRegression()                        #object for linear regression
predict_model.fit(data[['PM2.5','PM10','SO2','O3']],data['AQI'])         #model.fit(x,y)
```

```python
predict_model.predict([[88,135,28,115]])              #model.predict(x) 
```

```python
predict_model.score(data[['PM2.5','PM10','SO2','O3']],data['AQI'])   #model.score(x,y) for accuracy
```

```python
predict_model.coef_                                       #eqn : y = m1x1 + m2x2 + m3x3 + m4x4 + b
```

```python
predict_model.intercept_
```
